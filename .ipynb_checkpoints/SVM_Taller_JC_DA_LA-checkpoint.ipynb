```python
def gaussianKernel(x1, x2, sigma):
    # Ensure x1 and x2 are numpy arrays
    x1 = np.array(x1)
    x2 = np.array(x2)
    
    # Compute the squared Euclidean distance between x1 and x2
    diff = x1 - x2
    dist_squared = np.dot(diff, diff)
    
    # Compute the Gaussian kernel
    K = np.exp(-dist_squared / (2 * sigma**2))
    
    return K  
```

```python
x1 = -2.1
x2 = -1.2
sigma = 0.7;
K = gaussianKernel(x1, x2, sigma)
print('one dimension:', K)

x1 = np.array([1, 1])
x2 = np.array([1.5, 2]) 
sigma = 0.7;
K = gaussianKernel(x1, x2, sigma)
print('two dimension:', K)
```

Nota: valores más pequeños de 𝜎 hacen que el kernel decaiga más rápido, mientras que valores grandes mantienen la similitud entre puntos distantes.

Explore cómo afecta el valor de $\sigma$ cuando hay puntos cerca o muy separados.

```python

#Puntos cercanos
x1 = -2.1
x2 = -2.05

#Puntos lejanos
x3 = -2.1
x4 = -1.2

sigma_pequeño = 0.1
sigma_grande = 5

# Puntos cercanos (1D)
K_peq_1d = gaussianKernel(x1, x2, sigma_pequeño)
K_grande_1d = gaussianKernel(x1, x2, sigma_grande)

# Puntos lejanos (1D)
K_peq_1d_far = gaussianKernel(x3, x4, sigma_pequeño)
K_grande_1d_far = gaussianKernel(x3, x4, sigma_grande)

print(f'Kernel cercano (sigma pequeño): {K_peq_1d}')
print(f'Kernel cercano (sigma grande): {K_grande_1d}')
print(f'Kernel lejano (sigma pequeño): {K_peq_1d_far}')
print(f'Kernel lejano (sigma grande): {K_grande_1d_far}')
```

```python
x_vals = [x1, x2, x3, x4]
y_vals_peq = [K_peq_1d, K_peq_1d, K_peq_1d_far, K_peq_1d_far]  # Kernel con sigma pequeño
y_vals_grande = [K_grande_1d, K_grande_1d, K_grande_1d_far, K_grande_1d_far]  # Kernel con sigma grande

plt.figure(figsize=(10, 6))
plt.plot(x_vals[:2], y_vals_peq[:2], 'bo-', label='Puntos Cercanos (sigma pequeño)')
plt.plot(x_vals[2:], y_vals_peq[2:], 'ro-', label='Puntos Lejanos (sigma pequeño)')

plt.plot(x_vals[:2], y_vals_grande[:2], 'b--', label='Puntos Cercanos (sigma grande)')
plt.plot(x_vals[2:], y_vals_grande[2:], 'r--', label='Puntos Lejanos (sigma grande)')

plt.xlabel('Puntos en 1D')
plt.ylabel('Valor del Kernel')
plt.title('Efecto del parámetro sigma en el Kernel Gaussiano')
plt.legend()
plt.grid(True)
plt.show()
```

## **Interpretación** 
Se observa que para puntos cercanos, el 𝐾 es más grande con 𝜎 grande, ya que el valor del kernel se mantiene alto incluso si los puntos están un poco distantes.En cambio para puntos distantes (lejanos), 𝐾 es más pequeño con 𝜎 pequeño, reflejando la disminución en la similitud conforme aumenta la distancia.

*Nota:* El parámetro 𝐶 controla el nivel de penalización por clasificar incorrectamente ejemplos en los datos de entrenamiento.

*𝐶 grande:* Se penaliza mucho por los errores de clasificación. El SVM tratará de encontrar un hiperplano que clasifique correctamente todos los puntos, aunque esto puede llevar a un sobreajuste si los puntos son ruidosos o si hay ejemplos atípicos.

*𝐶 pequeño:* Se permite cierto número de errores de clasificación. El modelo tiende a ser más generalizable, pero puede no clasificar todos los ejemplos correctamente. Este enfoque puede generar un límite de decisión más suave.

```python

##Explrando con diferentes valores de C
data = np.loadtxt('dataset1.data', delimiter=',')
X = data[:, :2]
Y = data[:, 2].astype(int)

# Función para graficar el hiperplano
def plot_hyperplane(clf, min_x, max_x, linestyle, label):
    w = clf.coef_[0]
    a = -w[0] / w[1]
    xx = np.linspace(min_x - 2, max_x + 2)
    yy = a * xx - (clf.intercept_[0]) / w[1]
    plt.plot(xx, yy, linestyle, label=label)

# Graficar los datos
plt.figure(figsize=(8, 6))
plt.plot(X[Y == 0, 0], X[Y == 0, 1], '+r', label='Clase 0')
plt.plot(X[Y == 1, 0], X[Y == 1, 1], 'ob', label='Clase 1')

# Probar con diferentes valores de C
C_values = [1, 25, 100]

for C in C_values:
    clf_lin = svm.SVC(kernel='linear', C=C)
    clf_lin.fit(X, Y)
    
    # Graficar frontera
    min_x = X[:, 0].min()
    max_x = X[:, 0].max()
    plot_hyperplane(clf_lin, min_x, max_x, '--', label=f'C={C}')

plt.legend()
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Efecto de C en el límite de decisión de SVM')
plt.grid(True)
plt.show()
```

```python
### Revisar las metricas de clasificación para diferentes valores de C para ver que el modelo esta prediciendo correctamente
data = np.loadtxt('dataset1.data', delimiter=',')
X = data[:, :2]
Y = data[:, 2].astype(int)

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# Probar con diferentes valores de C
C_values = [0.1, 1, 10, 100]

for C in C_values:
    clf_lin = svm.SVC(kernel='linear', C=C)
    clf_lin.fit(X_train, Y_train)
    
    # Predecir las etiquetas para el conjunto de test
    Y_pred = clf_lin.predict(X_test)
    
    # Mostrar las métricas de clasificación
    print(f"\nResultados para C = {C}:")
    print(classification_report(Y_test, Y_pred))
    print("Matriz de confusión:")
    print(confusion_matrix(Y_test, Y_pred))

```

### Ejercicio 3

La siguiente parte cargará y trazará el conjunto de datos 2. En el gráfico, puede observar que no existe un límite de decisión lineal que separe los ejemplos positivos y negativos para este conjunto de datos. Sin embargo, al utilizar el núcleo gaussiano con SVM, podrá aprender un límite de decisión no lineal que puede funcionar razonablemente bien para el conjunto de datos. 

Debe cambiar el parámetro $\gamma$, está asociado al parámetro $\sigma$, de la primera ecuación. No debe utilizar valores superiores a 50.

¿Cuál es el efecto de cambiar el parámetro $\gamma$?

```python

########################################################################################
def plot_hyperplane(clf, X):
    # create a mesh to plot in
    h=0.01
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    # Put the result into a color plot
    Z = Z.reshape(xx.shape)

    CS=plt.contourf(xx, yy, Z,cmap=plt.cm.bone,alpha=0.5)
    #CS.cmap.set_under('yellow')
    #CS.cmap.set_over('cyan')

########################################################################################
#cargar dataset2
data=np.loadtxt('dataset2.data',delimiter=',')
X=data[:,:2]
Y=data[:,2].astype(int)
########################
##graficar datos
plt.figure(figsize=(8, 6))

plt.plot(X[Y==0,0],X[Y==0,1],'or')
plt.plot(X[Y==1,0],X[Y==1,1],'ob')
########################################################################
#create classifiers
C    = 1  # parametro de regularizacion de la svm
gamma= 1 #parametro gamma de la funcion de base radial
clf_rbf     = svm.SVC(kernel='rbf', gamma=gamma, C=C)
#entrenar
clf_rbf.fit(X, Y)

#graficar frontera
plot_hyperplane(clf_rbf,X)

```

## **Interpretación**

*Valor bajo de 𝛾:* Si 𝛾 es pequeño, el alcance de la influencia de cada punto de entrenamiento es grande. Esto significa que los puntos de entrenamiento "lejanos" tendrán una gran influencia en la clasificación, y el modelo podría ser más suave, lo que resulta en un límite de decisión más difuso.
Un valor bajo de 𝛾 tiende a subajustar el modelo, ya que el modelo no se adapta bien a los detalles de los datos y tiene un gran margen.

*Valor alto de 𝛾:* Si 𝛾 es grande, la influencia de los puntos cercanos es más localizada y su influencia se concentra en puntos cercanos, lo que permite que el modelo capture más detalles del conjunto de datos. Sin embargo, un valor alto de 𝛾 puede llevar a un sobreajuste porque el modelo se ajusta demasiado a los puntos de entrenamiento, creando un límite de decisión muy complejo que podría fallar en generalizar a nuevos datos.

```python
# Probando diferentes valores de gamma
gamma_values = [0.1, 1, 10, 50]

for gamma in gamma_values:
    clf_rbf = svm.SVC(kernel='rbf', gamma=gamma, C=1)
    clf_rbf.fit(X, Y)
    
    plt.figure(figsize=(8, 6))
    plt.title(f'RBF Kernel con gamma={gamma}')
    plot_hyperplane(clf_rbf, X)
    plt.scatter(X[Y == 0, 0], X[Y == 0, 1], color='red', label="Clase 0")
    plt.scatter(X[Y == 1, 0], X[Y == 1, 1], color='blue', label="Clase 1")
    plt.legend()
    plt.show()
```


𝛾 =0.1: El límite de decisión se observa más suave y abarca una mayor parte del espacio de características. Se evidencia que no se ajusta bien a las formas complejas de los datos.

𝛾=1:  El límite de decisión está más ajustado a los datos proporcionando un buen equilibrio entre la capacidad de generalización y el ajuste a los datos.

𝛾=50: Se observa que el límite de decisión es algo complejo y está muy bien ajustad a los puntos de entrenamiento, sin embargo se puede estar incurriendo en un sobreajuste del modelo.

```python
def plot_hyperplane(clf, X):
    h = 0.01
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    CS = plt.contourf(xx, yy, Z, cmap=plt.cm.bone, alpha=0.5)

# dataset de entrenamiento y prueba
data_train = np.loadtxt('dataset3_train.data', delimiter=',')
X = data_train[:, :2]
y = data_train[:, 2].astype(int)
data_test = np.loadtxt('dataset3_test.data', delimiter=',')
X_test = data_test[:, :2]
y_test = data_test[:, 2].astype(int)

# Graficar los datos de entrenamiento
plt.figure(figsize=(8, 6))
plt.plot(X[y == 0, 0], X[y == 0, 1], 'or')
plt.plot(X[y == 1, 0], X[y == 1, 1], 'ob')

# Definir los valores posibles de C y gamma
C = np.array([0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30])       
gamma = np.array([0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30])

# Inicializar variables para guardar el mejor modelo
best_accuracy = 0
best_C = None
best_gamma = None
best_model = None

# Generar combinaciones de C y gamma
for c in C:
    for g in gamma:
        clf_rbf = svm.SVC(kernel='rbf', gamma=g, C=c)
        clf_rbf.fit(X, y)
        y_pred = clf_rbf.predict(X_test)
        accuracy = (y_test == y_pred).sum() / y_test.size * 100
        print(f"Para C = {c} y gamma = {g}, la precisión es {accuracy:.2f}%")
        
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_C = c
            best_gamma = g
            best_model = clf_rbf

# Mostrar la mejor precisión y los parámetros seleccionados
print(f"\nEl mejor modelo tiene C = {best_C}, gamma = {best_gamma} y una precisión de {best_accuracy:.2f}%")

# Graficar la frontera de decisión para el mejor modelo
plt.figure(figsize=(8, 6))
plt.title(f'SVM con Kernel RBF (C={best_C}, gamma={best_gamma})')
plot_hyperplane(best_model, X)
plt.scatter(X[y == 0, 0], X[y == 0, 1], color='red', label='Clase 0')
plt.scatter(X[y == 1, 0], X[y == 1, 1], color='blue', label='Clase 1')
plt.legend()
plt.show()
```

Utilizando GridSearchCV

```python

def plot_hyperplane(clf, X):
    h = 0.01
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    CS = plt.contourf(xx, yy, Z, cmap=plt.cm.bone, alpha=0.5)

# Cargar el dataset de entrenamiento y prueba
data_train = np.loadtxt('dataset3_train.data', delimiter=',')
X = data_train[:, :2]
y = data_train[:, 2].astype(int)
data_test = np.loadtxt('dataset3_test.data', delimiter=',')
X_test = data_test[:, :2]
y_test = data_test[:, 2].astype(int)

# Graficar los datos de entrenamiento
plt.figure(figsize=(8, 6))
plt.plot(X[y == 0, 0], X[y == 0, 1], 'or')
plt.plot(X[y == 1, 0], X[y == 1, 1], 'ob')

# Definir los parámetros a buscar
param_grid = {
    'C': [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30],       
    'gamma': [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30]
}

# Crear el clasificador SVM con kernel RBF
svc = svm.SVC(kernel='rbf')

# Crear el objeto GridSearchCV
grid_search = GridSearchCV(svc, param_grid, cv=5, scoring='accuracy')

# Entrenar el modelo con GridSearchCV
grid_search.fit(X, y)

# Obtener el mejor modelo y los mejores parámetros
best_model = grid_search.best_estimator_
best_C = grid_search.best_params_['C']
best_gamma = grid_search.best_params_['gamma']

# Evaluar el modelo en el conjunto de prueba
y_pred = best_model.predict(X_test)
accuracy = (y_test == y_pred).sum() / y_test.size * 100
print(f"Mejores parámetros: C = {best_C}, gamma = {best_gamma}")
print(f"Precisión del mejor modelo en el conjunto de prueba: {accuracy:.2f}%")

# Graficar la frontera de decisión para el mejor modelo
plt.figure(figsize=(8, 6))
plt.title(f'SVM con Kernel RBF (C={best_C}, gamma={best_gamma})')
plot_hyperplane(best_model, X)
plt.scatter(X[y == 0, 0], X[y == 0, 1], color='red', label='Clase 0')
plt.scatter(X[y == 1, 0], X[y == 1, 1], color='blue', label='Clase 1')
plt.legend()
plt.show()
```
